```python
import matplotlib.pyplot as plt
Hours_of_study = [8,9,10,11,12]
grade_points = [7,8,9,9.5,10]
plt.bar(Hours_of_study,grade_points)
plt.title("Bar Chart")
plt.xlabel("Hours_of_study")
plt.ylabel("grade_points")
plt.show()
```

```python
import matplotlib.pyplot as plt
Hours_of_study = [8,9,10,11,12]
grade_points = [7,8,9,9.5,10]


plt.plot(Hours_of_study,grade_points)
plt.title("Line Plot")
plt.xlabel("Hours_of_study")
plt.ylabel("grade_points")
plt.show()
```

```python
import matplotlib.pyplot as plt
Hours_of_study = [8,9,10,11,12]
grade_points = [7,8,9,9.5,10]
plt.scatter(Hours_of_study,grade_points)
plt.title("Scatter Plot")
plt.xlabel("Hours_of_study")
plt.ylabel("grade_points")
plt.show()
```

```python
import matplotlib.pyplot as plt


grade_points = [5,6,8,9,10]

plt.hist(grade_points, bins = 3)
plt.title("Histogram")
plt.xlabel("Hours_of_study")
plt.ylabel("number of students")
plt.show()
```

```python
import matplotlib.pyplot as plt
grade_points = [5,6,8,9,10]

plt.boxplot(grade_points)
plt.title("Box plot")
plt.yticks(range(5, 11))
plt.ylabel("grade_points")
plt.show()
```

```python
import matplotlib.pyplot as plt


grade_points=[5,6,7,8,9]
mylabels = ["8 hrs","9 hrs","10 hrs","11 hrs","12 hrs"]

plt.pie(grade_points, labels = mylabels)
plt.show()
```
